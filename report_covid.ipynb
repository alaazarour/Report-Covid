```python
import pandas as pd
import numpy as np
pstore = pd.read_csv("covid_19_india.csv")
pstore.head(10)
```

```python
#importing all the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Create a distribution plot for rating
sns.distplot(pstore.Confirmed)
plt.show()
sns.distplot(pstore.Confirmed,bins=5, kde =True)
plt.show()
```

```python
#Create a distribution plot for rating
sns.distplot(pstore.Sno, bins=20, color="black")
plt.title("covid_19_india", fontsize=20, color = 'red')
plt.style.use("seaborn-whitegrid")
plt.show()

#Analyzing the Content Rating column
pstore['Sno'].value_counts()
```

```python
pstore = pstore[~pstore['Sno'].isin([9518,2049])]

#Resetting the index
pstore.reset_index(inplace=True, drop=True)

#Analyzing the Content Rating column
pstore['Sno'].value_counts()
```

```python
#importing all the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Plotting a bar chart
plt.figure(figsize=[9,7])
pstore['State/UnionTerritory'].value_counts().plot.pie()
plt.show()
```

```python
plt.style.use("dark_background")
plt.figure(figsize=[9,10])
pstore['State/UnionTerritory'].value_counts().plot.barh()
plt.show()
```
